```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
results_path = "results_grid_search_20103445/results.csv"
results = pd.read_csv(results_path, index_col=[0, 1, 2, 3], header=[0])
results["duration"] /= 1_000_000
results["remaining_vars"] = results["remaining_vars"].clip(0, 1)
N = 5
```

```python
if len(results[results["remaining_vars"] > 0]) == 0:
    results = results.drop(columns=["remaining_vars"])
else:
    print("======== SOME RUNS DID NOT FINISH ========")
```

```python
formulas = results.index.get_level_values("formula").unique()
complete = results.index.get_level_values("complete-minimization-relative-size").unique()
partial = results.index.get_level_values("partial-minimization-relative-size").unique()
incremental = results.index.get_level_values("incremental-absorption-removal-relative-size").unique()
print(formulas)
print(complete)
print(partial)
print(incremental)
```

```python
results = results.loc[(slice(None), complete[2], slice(None), slice(None))]
```

```python
def get_best(formula, n):
    filtered = results.loc[formula]
    return filtered.nsmallest(n, "duration")

def summarize_formula(formula):
    filtered = results.loc[formula]
    print(formula)
    display(filtered.nsmallest(N, "duration"))
    display(filtered.nlargest(N, "duration"))
```
